```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
input_directory = r'C:\Users\hp\Desktop\bmsis-ysp\Elemental Composition assignments (.csv)'
output_directory = r'C:\Users\hp\Desktop\bmsis-ysp\mass_spectrum_plots'
```

```python
os.makedirs(output_directory, exist_ok=True)
file_list = [f for f in os.listdir(input_directory) if f.endswith('.csv')]
```

```python
for filename in file_list:
    filepath = os.path.join(input_directory, filename)
    data = pd.read_csv(filepath, skiprows=2, usecols=[0, 1, 2, 5],
                       names=['#', 'Exp. m/z', 'Recal. m/z', 'Rel. Abun'])

    sample_name = filepath.split('_', 2)[2].split('Sum')[0]

    plt.figure(figsize=(6, 6))
    plt.bar(data['Recal. m/z'], data['Rel. Abun'], width=1)
    plt.minorticks_on()
    plt.title(f'Mass Spectrum Plot: {sample_name}')
    plt.xlabel('m/z')
    plt.ylabel('Signal Intensity')
    plt.tight_layout()

    output_filepath = os.path.join(output_directory, f'{sample_name}.png')
    plt.savefig(output_filepath)
    plt.show()
    plt.close()
```
