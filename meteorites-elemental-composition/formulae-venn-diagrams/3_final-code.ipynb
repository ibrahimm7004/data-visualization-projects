```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib_venn import venn2, venn3
from itertools import combinations
from upsetplot import UpSet
```

```python
df = pd.read_csv(r'c:\Users\hp\Desktop\bmsis-ysp\venn_diagrams\dataset_formulas.csv')
```

```python
df.head()
```

```python
def extract_unique_formulas(column):
    unique_formulas = set()
    for formula in column:
        unique_formulas.add(formula)
    return unique_formulas
```

```python
unique_formulas = {}
for col in df.columns:
    unique_formulas[col] = extract_unique_formulas(df[col])

# Prepare data for UpSet plot
unique_index = list(unique_formulas['2023March23NegESI_17_Mars_Meteorites126p8_Sum200.csv'])  # Using SampleA as the index
upset_data = pd.DataFrame(index=unique_index, columns=df.columns)

# Fill upset_data with boolean values
for col in df.columns:
    upset_data[col] = upset_data.index.isin(unique_formulas[col])

# Generate combinations of sets for intersections
combinations_list = list(combinations(df.columns, 2))

# Initialize UpSet plot correctly
upset = UpSet(upset_data, orientation='horizontal', sort_categories_by=None)

# Plot UpSet
upset.plot()
plt.title('UpSet Plot for Molecular Formulas')
plt.show()
```
